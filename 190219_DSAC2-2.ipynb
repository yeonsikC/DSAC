# example coding: calculating pi

```python
N = 10000
x = np.random.rand(N)
y = np.random.rand(N)

cnt = 0
for n in range(N):
    if ((x[n]*x[n]) + (y[n]*y[n])) < 1.0:
        cnt += 1
    else :
        continue
pi = 4*cnt/N
print(pi)

import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(5,5))       # plotting a circle
plt.scatter(x,y,s=1)             # s: marker a size

a = np.linspace(0,1,100)
b = (1-a*a)**0.5
plt.plot(a,b,c='r')
```

# 머신러닝

```python
# linear regression : y = a + bx

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

n = 100
x = np.random.randn(n) # 표준정규분포에서 100개 추출
y = 10 + 20*x       # a = 10 , b = 20
y = y + np.random.randn(n) * 10      # add noise
x,y

plt.figure(figsize=(5,5))
plt.scatter(x,y)

# -------------------------------------

b = np.random.randn()        # random initial choise
a = np.random.randn()

lr = 0.7                     # learning(or training) rate
n_epoch = 300    #최대 300번까지만 돌려라

for epoch in range(n_epoch):
    y_pred = a + b*x
    error = ((y_pred - y)**2).mean()
    if error < 1:
        break
    b = b - lr*((y_pred-y)*x).mean()
    a = a - lr*((y_pred-y)).mean()
    if epoch%10 == 0:
        print("epoch=", epoch, "error=", error, "a=", a, "b=", b)

print("------------------------")
print("epoch=", epoch, "error=", error, "a=", a, "b=", b)

xx = np.linspace(-5,5,100)
yy = a + b*xx
plt.plot(xx,yy,c='k')
```

```python
# linear regression : y = a + b1x1 + b2x2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

n = 100
x1 = np.random.randn(n)
x2 = np.random.randn(n)
y = 50 + 30*x1 + 40*x2       # a = 50 , b1 = 30, b2 = 40
y = y + np.random.randn(n) * 15      # add noise

b1 = np.random.randn()        # random initial choise
b2 = np.random.randn()
a = np.random.randn()

lr = 1.4                  # learning(or training) rate
n_epoch = 3000    #최대 3000번까지만 돌려라

for epoch in range(n_epoch):
    y_pred = a + b1*x1 + b2*x2
    error = ((y_pred - y)**2).mean() # MSE
    if error < 0.1:
        break
    b1 = b1 - lr*((y_pred-y)*x1).mean() # 미분.
    b2 = b2 - lr*((y_pred-y)*x2).mean()
    a = a - lr*((y_pred-y)).mean()
    if epoch%10 == 0:
        print("epoch=", epoch, "error=", error, "a=", a, "b1=", b1, "b2=", b2)

print("------------------------")
print("epoch=", epoch, "error=", error, "a=", a, "b1=", b1, "b2=", b2)
```

```python
# linear regression : y = a + b1x1 + b2x2 (function call)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

n = 100
x1 = np.random.randn(n)
x2 = np.random.randn(n)
y = 50 + 30*x1 + 40*x2       # a = 50 , b1 = 30, b2 = 40
y = y + np.random.randn(n) * 15      # add noise
```

```python
X = pd.DataFrame(x1)
X[1] = x2

from sklearn.linear_model import LinearRegression # sklearn라는 패키지 속, linear_model이라는 작은 패키지에서 LinearRegression 펑션을 가져옴.
model = LinearRegression() # 펑션
model.fit(X,y) # 모델 찾는것.
model.coef_, model.intercept_  # coef_ : b1, b2 # intercept_ : a
```

```python
new_X = [1,3]
model.predict([new_X])
```
